```python
# Part 1
with open("day3_input.txt", "r") as file:
    input = file.read().splitlines()

print(input[0:10])
```

```python
#Priorities
i = 1
priorities = {}
while i < 53:
    if i < 27:
        priorities[chr(i+96)] = i
    else:
        priorities[chr(i+38)] = i
    i = i + 1

print(priorities)
```

```python
sumOfPriorities = 0

for item in input:
    #print(item)
    compartiment1 = item[0:int(len(item)/2)]
    compartiment2 = item[int(len(item)/2):int(len(item))]
    for i in compartiment1:
        for y in compartiment2:
            if i == y: 
                sumOfPriorities += priorities.get(i)
                #print(i, priorities.get(i))
                break
        if i == y: break
    #print(compartiment1)
    #print(compartiment2)

print(sumOfPriorities)
```

```python
# Part 2
sumOfPriorities = 0
group = []

for item in input:
    group.append(item)
    if len(group) == 3:
        # Set récupère les charactère unique de chaque string puis récupère celui en commun entre les 3
        badge = set.intersection(set(group[0]),set(group[1]),set(group[2]))
        sumOfPriorities = sumOfPriorities +  priorities.get(list(badge)[0])
        group = []

print(sumOfPriorities)
```
